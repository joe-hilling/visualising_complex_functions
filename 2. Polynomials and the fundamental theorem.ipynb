```python
%matplotlib inline 
from domain_colourings import hsv1,rgb1,rgbLog,rgbReal
from plot_apparatus import ComplexPlot
from polynomials import Polynomial
import numpy as np
from numpy.random import random
import functions as fn

# Create a plotting object defaults to hsv
cPlot = ComplexPlot(hsv1)

# Domain Colourings
colArr = [hsv1, rgb1, rgbLog(10)]  
```

```python
n = 6
roots = 3*(random(n) + random(n)*1j -(.5 + .5j))
rndPoly = Polynomial(roots).function

cPlot.view(1.8)
cPlot.pointDensity = 200 # shows the more complex data
cPlot.plotMultiColorings('6 random poles', rndPoly, colArr)
```

```python
cPlot.plotMultiColorings('6 random poles inverted', lambda z: 1/rndPoly(z), colArr)
```

```python
cPlot.plot(lambda z: z,'Z or Identity')
```
